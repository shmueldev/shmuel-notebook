```python
# Pide al usuario un número n y muestra la suma de todos los números desde 1 hasta n.
numUser = int((input('Introduce un número')))
suma = 0
for n in range(numUser + 1):
    suma += n
    print(suma)
```

```python
# Muestra todos los números pares entre 1 y 50.
for n in range(51):
    if n % 2 == 0:
        print(n)
```

```python
# Pide un número al usuario y muestra su tabla de multiplicar del 1 al 10.
numUser = int(input('Ingrese un número'))
for n in range(11):
    tabla = n * numUser
    print(tabla)
```

```python
# Genera un número aleatorio entre 1 y 3 y permite al usuario adivinarlo.
# Debes dar pistas: “más alto” o “más bajo”.
import random
numero = random.randint(1, 3)
numUser = int(input('Ingrese un número para adivinar el número aleatorio'))
intentos = 0

while numUser != numero:
    if numUser > numero:
        print(f'el número que pusiste es mayor {numUser}')
    else:
        print(f'el número que pusiste es menor {numUser}')

    numUser = int(input('Ingrese un número para adivinar el número aleatorio'))
    intentos += 1

print(f"El número que escribiste es {numUser} correcto, el número de intentos es {intentos + 1}")
```

```python
# Pide una palabra o frase y cuenta cuántas vocales tiene.
textUser = (input('Ingrese una frase')).lower()
vocales = ["a","e","i","o","u" ]
contador_vocales = 0

for l in textUser:
    if l in vocales:
        contador_vocales += 1

print(contador_vocales)
```

```python
# Pide un número y determina si es primo o no.
numUser = int(input('Ingrese un número'))
es_primo = True

if numUser <= 1:
    es_primo = False
    print(f'El número que escribiste no es primo {numUser}')
else:
    for n in range(2, numUser):
        if numUser % n == 0:
            es_primo = False
            break
    print(f'El número que escribiste es {es_primo}')
```

```python
# Calcula el factorial de un número ingresado por el usuario.
numUser = int(input('Ingrese un número'))
acomulador = 1

for n in range(1, numUser + 1):
    acomulador = acomulador * n
    print(acomulador)
```
