```python
## Eficiencia en el proccesamiento de datos
datos = [
    {"cliente": "Juan", "ventas": 100, "activo": True},
    {"cliente": "Samuel", "ventas": 200, "activo": False},
    {"cliente": "David", "ventas": 300, "activo": True},
    {"cliente": "Jose", "ventas": 200, "activo": False},
]

activos = []
for d in datos:
    if d["activo"] == True:
        activos.append(d)

ventas = []
for a in activos:
    ventas.append(a["ventas"])

total = 0
for v in ventas:
    total += v

# Muestra el total final de ventas por clientes activos
print("total", total)
```

```python
# Código eficiente
datos = [
    {"cliente": "Juan", "ventas": 100, "activo": True},
    {"cliente": "Samuel", "ventas": 200, "activo": False},
    {"cliente": "David", "ventas": 300, "activo": True},
    {"cliente": "Jose", "ventas": 200, "activo": False},
]

total = sum(d["ventas"] for d in datos if d["activo"])
print('total', total)
```

```python
datos = {
    'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado'],
    'Precio': [1200, 25, 300, 50],
    'Ventas': [5, 50, 10, 20]
}

# Convertimos a un DataFrame
df = pd.DataFrame(datos)

# Calculamos una nueva columna
df['Total_Ventas'] = df['Precio'] * df['Ventas']

# Filtramos productos caros (más de 100)
caros = df[df['Precio'] > 100]

print(caros)
```
