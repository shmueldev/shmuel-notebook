```python
# Cargar datos
df = pd.read_csv('Titanic-Dataset.csv')
print(df)

# Usamos .copy() al final para que sea un objeto totalmente independiente
df_limpio = df.dropna().copy()
# Reajustas los índices (opcional pero recomendado)
df_limpio = df_limpio.reset_index(drop=True)

print(f"Filas originales: {len(df)}")
print(f"Filas después de limpiar: {len(df_limpio)}")
```

```python
# Limpieza
df.isnull().sum()

# Vamos a rellenar los campos nulos de age con la edad promedio
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# ¿Qué pasajeros tienen mayor tasa de supervivencia?
df.groupby('Sex')['Survived'].mean()
```

```python
# ¿Qué variable impacta más: clase, género o edad?
df.groupby('Sex')['Survived'].mean()
```

```python
# ¿A quién priorizarías en la evacuación?
(df['Sex'] == 'female') | (df['Age'] < 12)
```

```python
# Prioridad compuesta
def prioridad(row):
    if row['Sex'] == 'female' or row['Age'] < 18:
        return 1
    elif row['Pclass'] == 3:
        return 2
    else:
        return 3

df['Priority'] = df.apply(prioridad, axis=1)
df.head()
```
